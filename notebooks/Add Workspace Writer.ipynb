```python
## Filter out workspaces without write access
workspaces = FAPI.list_workspaces().json()

pp.pprint(workspaces[0])

workspaceNames = set()
for w in workspaces:
    if w['workspace']['namespace'] == "anvil-datastorage" and w['workspace']['name'] in workspaceTags.index:
        workspaceNames.add(w['workspace']['name'])



reader = set()
writer = set()
for w in workspaces:
    name = w['workspace']['name']
    if name in workspaceNames:
        if w['accessLevel'] == "READER":
            reader.add(name)
        if w['accessLevel'] == "WRITER" or w['accessLevel'] == "OWNER":
            writer.add(name)

        

print(len(workspaces))
print(len(workspaceNames))
print(len(workspaceTags.index))
print(len(reader))
print(len(writer))
pp.pprint(reader)
```
